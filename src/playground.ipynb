```python
import datamanager as dmng
import matplotlib.pyplot as plt

dm = dmng.DataManager(read_mode='feature')

```

```python
%matplotlib qt
import numpy as np
import pandas as pd

data = dm.data['AT']

fig, ax = plt.subplots()


time = data['time']
price = data['price']
data_ti = pd.DataFrame(data).set_index('time')
price_ti = data_ti['price']



price_ti_ma = price_ti.rolling(window='24h', center=True).mean()

ax.plot(time, price)
ax.plot(time, price_ti_ma)

plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 1. Define the Neural Network
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        return out

# 2. Hyperparameters
input_size = 10
hidden_size = 64
output_size = 1
learning_rate = 0.001
batch_size = 32
num_epochs = 10

# 3. Create Dummy Data (Replace with your actual data loading)
# X_train shape: (num_samples, input_size)
# y_train shape: (num_samples, output_size)
X_train = torch.randn(1000, input_size)
y_train = torch.randn(1000, output_size)

dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(dataset=dataset, batch_size=batch_size, shuffle=True)

# 4. Initialize Model, Loss, and Optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleNN(input_size, hidden_size, output_size).to(device)

criterion = nn.MSELoss() # Use nn.CrossEntropyLoss() for classification
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 5. Training Loop
for epoch in range(num_epochs):
    model.train() # Set model to training mode
    running_loss = 0.0
    
    for i, (inputs, targets) in enumerate(train_loader):
        inputs = inputs.to(device)
        targets = targets.to(device)
        
        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(train_loader):.4f}')

# 6. Evaluation (Example)
model.eval() # Set model to evaluation mode
with torch.no_grad():import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 1. Define the Neural Network
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        return out

# 2. Hyperparameters
input_size = 10
hidden_size = 64
output_size = 1
learning_rate = 0.001
batch_size = 32
num_epochs = 10

# 3. Create Dummy Data (Replace with your actual data loading)
# X_train shape: (num_samples, input_size)
# y_train shape: (num_samples, output_size)
X_train = torch.randn(1000, input_size)
y_train = torch.randn(1000, output_size)

dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(dataset=dataset, batch_size=batch_size, shuffle=True)

# 4. Initialize Model, Loss, and Optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleNN(input_size, hidden_size, output_size).to(device)

criterion = nn.MSELoss() # Use nn.CrossEntropyLoss() for classification
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 5. Training Loop
for epoch in range(num_epochs):
    model.train() # Set model to training mode
    running_loss = 0.0
    
    for i, (inputs, targets) in enumerate(train_loader):
        inputs = inputs.to(device)
        targets = targets.to(device)
        
        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(train_loader):.4f}')

# 6. Evaluation (Example)
model.eval() # Set model to evaluation mode
with torch.no_grad():
```

```python
# Correlation and covariance

prc = np.array(price_ti[-11:])

fig, ax = plt.subplots()
t_i = prc[1:]
t_im1 = prc[:-1]
ax.plot(t_i, '.-', label=r'$T_i$')
ax.plot(t_im1, '.-', label=r'$T_{i-1}$')
ax.legend()

t_i_mean = np.mean(t_i)
t_im1_mean = np.mean(t_im1)
t_i_std = np.std(t_i)
t_im1_std = np.std(t_im1)
n = len(t_i)

cov = 1/n * np.sum((t_i-t_i_mean)*(t_im1-t_im1_mean))
print('Covariant', cov)
cor = cov / (t_i_std*t_im1_std)
print('Correlation', cor)
cor_np = np.corrcoef(t_i, t_im1)
print('Correlation (numpy)', cor_np[0][1])

fig, ax = plt.subplots()
ax.scatter(t_i, t_im1)
fit_model = np.polyfit(t_i, t_im1, 1)

fit = np.poly1d(fit_model)
fit_tm1 = fit(t_i)
ax.plot(t_i, fit_tm1, color='red')
```
